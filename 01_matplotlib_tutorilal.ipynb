# Anatomy of a plot

```python
# 2 main components in a plot, figure and axis
import matplotlib.pyplot as plt
%matplotlib inline
#plt.show() when running from a source script
```

```python
# 1. functional approach-using basic matplotlib commands
import numpy as np
```

```python
#generate 20 no's between 0 & 10
X=np.linspace(0,10,20)
Y=X**2
```

```python
X
```

```python
Y
```

```python
plt.plot(X,Y)
```

```python
# adding title and labeling the axeses
plt.plot(X,Y)
plt.title('my first plot!'.upper())
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#for creating multiplots use .subplot
#plt.subplot(no of rows for figure,figure columns,plot no in the figure)
# plotting both the plots using subplot
plt.subplot(1,2,1)
plt.plot(X,Y,'green')


plt.subplot(1,2,2)
plt.plot(Y,X,'red');
```

```python
#2. object oriented interface: create figure objects and use methods of it:
# (left, bottom, width, and heightwhich are the positions where the axes should be placed)
# figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.set_title('2nd plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.plot(X,Y,'purple')
```

```python
# creating 2 figures using above approach
fig1=plt.figure(figsize=(5,6))
ax1=fig1.add_axes([1,1,1,1])
ax2=fig1.add_axes([1.2,1.2,0.5,0.5])
ax1.plot(X,Y,'blue')
ax2.plot(X,Y,'green')
ax1.set_title('plot 1')
ax2.set_title('plot 2')
```

```python
#plotting x and y on subplot(0,1) & subplot(1,2) respectively:

fig,ax=plt.subplots(nrows=3,ncols=3)
ax[0,1].plot(X,Y,'green')
ax[1,2].plot(Y,X,'red')
ax[0,1].set_title('subplot1')
ax[1,2].set_title('subplot2')
ax[0,1].set_xlabel('x-axis')
ax[0,1].set_ylabel('y-axis')
ax[1,2].set_xlabel('x-axis')
ax[1,2].set_ylabel('y-axis')
plt.tight_layout()
```

# Figure size,aspect ratio and DPI in matplot lib


### Matplotlib allows us create customized plots by specifying the figure size, aspect ratio, and DPI by simply specifying the figsize and dpi arguments. The figsize is a tuple of the width and height of the figure (in inches), and dpi is the dots-per-inch (pixel-per-inch).

```python
# resolution- 100
fig=plt.figure(figsize=(6,4),dpi=100)
ax=fig.add_axes([1,1,1,1])
ax.plot(X,Y)
```

```python
# same with subplots
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(8,4),dpi=100)
ax[0].plot(X,Y,'green')
ax[1].plot(Y,X,'blue')

plt.tight_layout()
```

```python
#saving plots
# figure is our image, we can save it in a number of formats, such as png, jpg, svg, pdf, etc
fig.savefig('matplotlib_tut.png')
```

# Legends and Customization

###  1. LEGENDS:  To identify the plots, we need to add a legend using .legend() and then specify the label=  attribute for each plot

```python


fig=plt.figure(figsize=(8,6),dpi=80)
ax=fig.add_axes([0,0,1,1])
ax.plot(X,X**2,'red',label='square')
ax.plot(X,X**3,label='cube')

ax.legend()
```

# 2. Plot Appearance- changing linewidth or lw , linestyle or ls, and mark out data points using marker

```python
fig=plt.figure(figsize=(8,6),dpi=80)
ax=fig.add_axes([0,0,1,1])
ax.plot(X,Y,'purple',linewidth=3,linestyle='--',marker='o',markersize=5)
```

# 3. Plot range- Matplotlib allows us to set limits for our plots. Setting thresholds for x and y axis of our plot

```python

fig=plt.figure(figsize=(8,6),dpi=60)
ax=fig.add_axes([0,0,1,1])
ax.plot(X,Y,lw=3,ls='--')

ax.set_xlim([0,1]) # upper and lower bound limit of x axis
ax.set_ylim([0,5]) # upper and lower bound limit of y axis
```
